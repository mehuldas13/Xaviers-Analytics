```python
# function for making ngrams
import nltk, re, string, collections
from collections import Counter
from nltk.util import ngrams
from nltk import word_tokenize


# this corpus is pretty big, so let's look at just one of the files in it
with open("Macbeth.txt", "r", encoding='utf-16-le') as file:
    text = file.read()  
   
```

```python
#preprocessing
text = re.sub('<.*>','',text)
text = re.sub('\n','',text)
text = re.sub('\t','',text)
text = re.sub('ENDOFARTICLE.','',text)
punctuationNoPeriod = "[" + re.sub("\.","",string.punctuation) + "]"
text = re.sub(punctuationNoPeriod, "", text)
text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
text = text.lower()
```

```python
text
```

```python
#Tokenizing and removing empty sapces in text
tokenized = text.split()
tokens = [token for token in text.split(" ") if token != ""]
```

```python
#Calculating unigram
unigram = ngrams(tokenized, n=1)
list(unigram)
```

```python
#Calculating bigram
bigram = ngrams(tokenized, n=2)
list(bigram)
```

```python
wordfreq = [tokenized.count(w) for w in tokenized] # a list comprehension
def wordListToFreqDict(tokenized):
    wordfreq = [tokenized.count(p) for p in tokenized]
    return dict(list(zip(tokenized,wordfreq)))
```

```python
print(dict(list(zip(tokenized,wordfreq))))
```

```python
#Frequency of unigram
n_gram = 1
unifreq=Counter(ngrams(tokenized, n_gram))
```

```python
#Frequency of bigram
n_gram = 2
bifreq=Counter(ngrams(tokenized, n_gram))
```

```python
text_length=len(text)
```

```python
# Number of unigram,bigram words
print(len(unifreq))
print(len(bifreq))

# Probability of Unigram words
for key in unifreq:
    unifreq[key] /= text_length
    Uniprob=unifreq
    
print(Uniprob)
print(len(Uniprob))
```

```python
# Probability of Bigram words
for unifreq in bifreq:
    bifreq[unifreq] /= text_length
    Biprob=bifreq
    
print(Biprob)
print(len(Biprob))
```
